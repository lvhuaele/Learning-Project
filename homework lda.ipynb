```python
def weekday(X):
    if X == 'wed':
        return 'wednesday'
    elif X == 'thur':
        return 'thursday'
    elif X == 'thurday':
        return 'thursday'
    elif X == 'fri':
        return 'friday'
    else:
        return X
```

```python
df.x35 = df.x35.apply(weekday)
```

```python
"""transform to numerical features"""
df.x41 = df_object.x41.map(lambda x: x if type(x) == float else float(x.strip('$')))
df.x45 = df_object.x45.map(lambda x: x if type(x) == float else float(x.strip('%')))
```

```python
"""split training and test data"""
y = df.y
X = df.drop(columns = ['y'])
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state= 50)
```

```python
for column in ['x34', 'x35', 'x68', 'x93']:
    X_test[column].fillna(X_train[column].mode()[0], inplace=True)
```

```python
"""fill missing values for all numerical variables"""
import numpy as np

for col in X_train.select_dtypes(include=np.number):
    X_train[col] = X_train[col].fillna(X_train[col].mean())
```

```python
for col in X_test.select_dtypes(include=np.number):
    X_test[col] = X_test[col].fillna(X_train[col].mean())
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
X_lda = lda.fit(X_train_scaled, y_train)
```

```python
lda_var_ratios = lda.explained_variance_ratio_
```

```python
# Create a function
def select_n_components(var_ratio, goal_var: float) -> int:
    # Set initial variance explained so far
    total_variance = 0.0
    
    # Set initial number of features
    n_components = 0
    
    # For the explained variance of each feature:
    for explained_variance in var_ratio:
        
        # Add the explained variance to the total
        total_variance += explained_variance
        
        # Add one to the number of components
        n_components += 1
        
        # If we reach our goal level of explained variance
        if total_variance >= goal_var:
            # End the loop
            break
            
    # Return the number of components
    return n_components
```

```python
select_n_components(lda_var_ratios, 0.95)
```
